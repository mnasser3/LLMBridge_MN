```python
model_name = "mistralai/Mistral-7B-v0.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the model with 8-bit precision and device map
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_8bit=True,
    device_map="auto"
)
```

```python
def generate_text(prompt, max_new_tokens=10):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(inputs.input_ids.to('cuda'), max_new_tokens=max_new_tokens, num_return_sequences=1, top_p=0.1)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
prompt = """does the word "ex" describe the focus of a workout? the answer is either yes or no"""

# Generate and print the output
generated_text = generate_text(prompt)
print(generated_text)
```
